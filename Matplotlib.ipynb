## Python' da Matplotlib Kütüphanesi
Matplotlib, Python'da kullanılan en popüler veri görselleştirme kütüphanelerinden biridir. Matplotlib, grafikler, çizimler ve görsel sunumlar oluşturmak için kullanılır ve birçok farklı grafik türünü destekler.

## boxplot(X)
Box plotlar, veri setinin merkezi eğilimini, dağılımını ve aykırı değerlerini hızlı bir şekilde anlamak için oldukça kullanışlıdır.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

### Pasta grafiği (pie chart)
Pasta grafikleri genellikle bir toplamın parçalarının ne kadarını temsil ettiğini veya bir kategorinin içindeki alt kategorilerin oranlarını göstermek için kullanılır. Ayrıca, bir veri setinin dağılımını anlamak ve farklı kategoriler arasındaki ilişkiyi görselleştirmek için de yararlı olabilir.

```python
import matplotlib.pyplot as plt

# Veri seti
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Renkler
explode = (0.1, 0, 0, 0.1)  # Dilimlerin dışarı doğru ayrılma miktarı

# Pasta grafiğini çiz
plt.figure(figsize=(4, 3))  # Grafiğin boyutu
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode)

# Başlık ve gölgenin eklenmesi
plt.title('Pasta Grafiği')
plt.axis('equal')

# Grafiği göster
plt.show()
```

## Streamplot()
Streamplot() işlevi, akış hattı grafiği veya akış hattı görselleştirmesi oluşturmak için kullanılır. Bu tür grafikler, bir akışın veya hareketin yönünü ve hızını 2D bir alanda görselleştirmek için kullanılır. Özellikle akışkanlar dinamiği, elektromanyetizma veya rüzgar gibi alanlarda kullanışlıdır.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# akış oluşturma
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# U ve V nin akıştan çıkarılması.
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

## 3D Scatter
matplotlib içindeki scatter() işlevi, veri noktalarını bir grafik üzerinde dağıtmak ve her noktanın konumunu ve belirli bir özellikle ilişkisini görsel olarak göstermek için kullanılır. Scatter plotlar (nokta grafikleri), veri setindeki her bir öğenin iki veya daha fazla değişken arasındaki ilişkiyi incelemek için kullanılır.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

### Animasyon

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots()
t = np.linspace(0, 3, 40)
g = -9.81
v0 = 12
z = g * t**2 / 2 + v0 * t

v02 = 5
z2 = g * t**2 / 2 + v02 * t

scat = ax.scatter(t[0], z[0], c="b", s=5, label=f'v0 = {v0} m/s')
line2 = ax.plot(t[0], z2[0], label=f'v0 = {v02} m/s')[0]
ax.set(xlim=[0, 3], ylim=[-4, 10], xlabel='Time [s]', ylabel='Z [m]')
ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    x = t[:frame]
    y = z[:frame]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # update the line plot:
    line2.set_xdata(t[:frame])
    line2.set_ydata(z2[:frame])
    return (scat, line2)


ani = animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)
plt.show()
```
