### Seaborn, Matplotlib ve Bokeh Arasındaki Farklar

**Matplotlib**:
- *Temel Çizim Kütüphanesi*: Matplotlib, Python için temel bir çizim kütüphanesidir. Diğer görselleştirme kütüphaneleri, Matplotlib üzerine inşa edilmiştir.
- *Düşük Seviyeli Kontrol*: Matplotlib, figürler, eksenler ve çizimler gibi düşük seviyeli kontrol sağlar, bu da özelleştirilebilirlik açısından büyük bir esneklik sunar.
- *Grafik Özelleştirme*: Tamamen özelleştirilebilir grafikler oluşturmak için kullanılabilir, ancak bazen daha fazla kod yazma gerekebilir.
- *Statik Grafikler*: Matplotlib ile oluşturulan grafikler genellikle statiktir ve görselleştirmeleri paylaşmak için görüntüler veya PDF dosyaları olarak kaydedilir.

**Seaborn**:
- *İstatistiksel Görselleştirme Odaklı*: Seaborn, istatistiksel analiz ve görselleştirmeler için özellikle tasarlanmıştır. İstatistiksel grafiklerin oluşturulmasını kolaylaştırmak için özel işlevler sunar.
- *Görsel Stil*: Seaborn, daha çekici varsayılan görsel stiller sunar ve grafiklerinizi daha hızlı ve kolay bir şekilde özelleştirmenize yardımcı olur.
- *Pandas Entegrasyonu*: Pandas DataFrame'leri ile uyumlu çalışır, bu da veri analizi ve görselleştirme işlemlerini entegre etmeyi kolaylaştırır.
- *İnteraktif Değil*: Seaborn, temel olarak istatistiksel analizler için kullanılır ve interaktif görselleştirmeleri desteklemez.

**Bokeh**:
- *İnteraktif Görselleştirme*: Bokeh, interaktif web tabanlı görselleştirmeler oluşturmak için kullanılır. Web tarayıcısında çalışan grafikler ve arayüzler oluşturabilir.
- *Veri Büyüklüğü*: Büyük veri kümeleriyle çalışırken iyi performans gösterir ve büyük verileri etkileşimli olarak keşfetmeyi sağlar.
- *Web Entegrasyonu*: Bokeh, web uygulamalarına ve panolarına (dashboards) görselleştirme entegre etmek için yaygın olarak kullanılır.
- *Daha Yüksek Seviyeli Araçlar*: Bokeh, Matplotlib ve Seaborn'a göre daha yüksek seviyeli araçlar sunar ve daha karmaşık grafikler oluşturmayı kolaylaştırır.


## Seaborn Kütüphanesi
Seaborn, Matplotlib kütüphanesine yüksek seviye arayüz sağlayan bir kütüphanedir. Seaborn ile;

* Estetik olarak hoş olan varsayılan temaları kullanma
* Özel olarak renk paleti belirleme
* Göz alıcı istatistiksel grafikler yapma
* Dağılımları kolayca ve esnek bir şekilde sergileme
* Matris ve DataFrame içindeki bilgileri görselleştirme;

yapılabilir. Yukarıdaki son üç madde, Seaborn’un Keşifçi Analiz için biçilmiş kaftan olduğunu gösteriyor. Veriniz hakkında hızlı ve etkili bir biçimde bilgi sahibi olabiliyorsunuz.

>Pokemon Veri seti üzerinden `seaboen` incelemesi

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Pokemon.csv", index_col=0)
df.head()
```

### Seaborn Grafik Fonksiyonları

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

>`lmplot()` aslında regresyon çizgisi için kullanılır fakat burada saçılım grafiği yapmak için kulanabiliriz.

```python
sns.lmplot(x = 'Attack', y = 'Defense', fit_reg = False, hue = 'Stage', data = df)
```

### Seaborn Temaları
Seaborn’un diğer bir avantajı ise havalı temalar ile beraber geliyor olmasıdır. Seaborn’un varsayılan teması darkgrid temasıdır.

>Pokémon’ların saldırı değerlerinin temel tiplerine göre dağılımlarını gösterelim.

```python
sns.set_style('whitegrid')
sns.violinplot(x = 'Type 1', y = 'Attack', data = df)
```

>Görüldüğü gibi ejderhaların hayaletlere göre saldırı değerleri daha yüksek, ancak hayaletlerin varyansı daha büyüktür.

**Renk Paletleri**

Seaborn'da kendi renk paletlerimizi oluşturabiliyoruz.

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x='Type 1', y='Attack', data=df,
               palette=pkmn_type_colors)
```

**swarm (sürü) grafiği**

Bu grafik bize her bir noktayı gösteriyor, ayrıca benzer noktaları birlikte gösteriyor.

```python
sns.swarmplot(x = 'Type 1', y = 'Attack', data = df, palette=pkmn_type_colors)
```

### Grafikleri Üstüste Bindirme


```python
sns.violinplot(x = 'Type 1',
               y = 'Attack',
               data = df,
               inner = None,
               palette=pkmn_type_colors)

sns.swarmplot(x = 'Type 1',
              y = 'Attack',
              data = df,
              color = 'k',
              alpha = 0.7)

plt.title('Attack by Type')
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis = 1)
sns.boxplot(data = stats_df, notch = True, linewidth = 2.5, width = 0.50)
```

###  Isı Haritası
Isı haritaları, matris tarzında verileri görselleştirmek için kullanılır.

```python
corr = stats_df.corr()

sns.heatmap(corr)
```

### Histogram
Histogram, nümerik değişkenlerin dağılımlarını görselleştirmede kullanılır.

```python
sns.distplot(df.Attack)
```

### Sütun Grafiği
Sütun grafiği, kategorik değişkenleri görselleştirmede kullanılır.

```python
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)

# X eksenindeki etiketleri döndür
plt.xticks(rotation=-45)
```

### Faktör Grafiği
Faktör grafiği, faktör sınıflarının grafiklerini kolayca ayırmaya yarar.

```python
g = sns.catplot (x = 'Type 1',
                   y = 'Attack',
                   data = df,
                   hue = 'Stage',
                   col = 'Stage',
                   kind = 'swarm')
g.set_xticklabels(rotation = -45)
```

### Yoğunluk Grafiği
Yoğunluk grafiği, iki değişken arasındaki dağılımı gösterir.

```python
sns.kdeplot(df.Attack, df.Defense)
```
