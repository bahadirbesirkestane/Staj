>**Resampling**

```python
import pandas as pd
import numpy as np
```

```python
# Verisetini okuyoruz ve date değikeninin tipini datetime yapıyoruz.
# index_col ile indexleri tarih yapısına çevirdik
fb=pd.read_csv("/content/META.csv",parse_dates=["Date"],index_col="Date")
fb.head()
```

```python
# Aylık ortalama değerler
fb.resample("M").mean()
```

```python
%matplotlib inline
fb.Close.resample("M").mean()
```

```python
# Aylık Ortalama değerlerin grafiği
fb.Close.resample("M").mean().plot()
```

```python
#Yılın 4 çeyerği için
fb.Close.resample("Q").mean().plot()
```

```python
# Bar grafik şeklinde görüntülemek için
fb.Close.resample("M").mean().plot(kind="bar")
```

>**Shifting**
>
> Zaman boyunca ileri ve geri gitmek

```python
fb1=pd.DataFrame(fb.Close["2023/03"])
fb1.head()
```

```python
# değerleri 2 zaman ileri ötelenmesi.
fb1.shift(2)
```

```python
fb1["Bir Önceki Fiyat"]=fb1.shift(1)
fb1
```

```python
# Değişimlerin Yüzdesinin bulunması
fb1["Bir Günük Değişim"]=fb1["Close"]-fb1["Bir Önceki Fiyat"]
fb1.head()
```

```python
# 1 Günlük değişimin yüzdesi
fb1["Yüzdelik Değişim"]=(fb1["Close"]-fb1["Bir Önceki Fiyat"])*100/fb1["Bir Önceki Fiyat"]
fb1.head()
```

>t-Shift

```python
fb2=fb1[["Close"]]
fb2.head()
```

```python
fb2.index
```

```python
# Freq=None atanmamış

fb2.index=pd.date_range("2023-03-01",periods=23,freq="B")
fb2.index
```

```python
# İndex lerle beraber öteleme
fb2.tshift(1)
```

```python
# 2 tarih geriye öteleme
fb2.tshift(-2)
```

```python
x=pd.date_range("12/9/2009 9:30",periods=6,freq="D")
zs=pd.Series(np.random.randn(len(x)),index=x)
zs
```

```python
print(zs.index.tz)
```

```python
zs_utc=zs.tz_localize("UTC")
zs_utc
```

```python
# Zaman Serisinin Belirli Bölgelere göre düzenlenmesi
zs_utc.tz_convert("US/Hawaii")
```

```python
# Bir zaman damgalı tarihi ulusararası zaman bölgesine dönüştürme
zdamga=pd.Timestamp("2023-06-26 05:00")
zdamga
```

```python
zdamga_utc=zdamga.tz_localize("utc")
zdamga_utc
```

```python
# Bu değeri istanbul yerel saatine dönüştürelim
zdamga_utc.tz_convert("Europe/Istanbul")
```
