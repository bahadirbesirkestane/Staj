### **Google Play Store Veri Seti ile Veri Temizleme ve Görselleştirme**

>Pandas Ktüphanesinin import edilmesi ve veri setinin okunması

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/googleplaystore.csv')
```

```python
df.head()
```

```python
df.columns
```

>Kolon isimlerinde boşluk karakterleri var. Bunları alt tire "_" ile değiştiricez.

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df.shape
```

```python
df.dtypes
```

>Bazı sütunların işlem yapabilmek için veri tiplerini değiştireceğiz.

>Eksik veriler

```python
df.isnull().sum()
```

>Eksik verileri Görsel Halinde görüntüleyelim

```python
import seaborn as sns
```

```python
sns.set_theme()
sns.set(rc={"figure.dpi":300,"figure.figsize":(12,9)})
```

```python
sns.heatmap(df.isnull(),cbar=False)
```

>Rating sütünundaki verilerin medyanını alarak boş hücrelere yazacağız.

```python
rating_medyan=df["Rating"].median()
rating_medyan
```

```python
df["Rating"].fillna(rating_medyan,inplace=True)
```

>Kalan eksik veriler az olduğu için kaldırılabilir.

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

>Belirli sütunların veri tiplerini değiştiriyoruz.

```python
df["Reviews"].describe()
```

```python
df["Reviews"]=df["Reviews"].astype("int64")
df["Reviews"].describe().round()
```

---

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

>Değerlerde k, m harfleri var ve String ifadeler var.

```python
df["Size"].replace("M","",regex=True,inplace=True)
df["Size"].replace("k","",regex=True,inplace=True)
```

```python
df["Size"].unique()
```

```python
size_median=df[df["Size"]!="Varies with device"]["Size"].astype(float).median()
size_median
```

>String ifade yerine sütunun medyanını yazıyoruz.

```python
df["Size"].replace("Varies with device",size_median,inplace=True)
```

```python
df["Size"]=pd.to_numeric(df["Size"])
```

```python
df["Size"].head()
```

```python
df.Size.describe().round()
```

>Install Sütunu

```python
df.Installs.unique()
```

>Virgüller ve "+" sembollerini kaldırmamız gerekiyor.

```python
df.Installs=df.Installs.apply(lambda x:x.replace("+",""))
df.Installs=df.Installs.apply(lambda x:x.replace(",",""))
df.Installs=df.Installs.apply(lambda x:int(x))
```

```python
df.Installs.unique()
```

>Price Sütunu

```python
df["Price"].unique()
```

```python
df.Price=df.Price.apply(lambda x:x.replace("$",""))
df.Price=df.Price.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

>Genres Sütunu

```python
len(df["Genres"].unique())
```

```python
df.Genres.head(10)
```

>Noktali virgül değerlerini kaldırmamız gerekiyor.

```python
df.Genres=df.Genres.str.split(";").str[0]
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].unique()
```

```python
df["Genres"].value_counts()
```

```python
df.Genres.replace("Music & Audio","Music",inplace=True)
```

>Last_Updated Sütunu

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"]=pd.to_datetime(df["Last_Updated"])
```

Bu işlemle beraber veri temizleme ve veri önişleme işlmeleri yapıldı.
